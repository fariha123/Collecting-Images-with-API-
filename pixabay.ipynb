```python
import os 
import requests

def find_pages_limit(terms):
    r = requests.get(f'https://pixabay.com/api/?key=21967882-1b9779f8b2e624ded8ba912a1&?page=1&q={terms}&image_type=photo')
    data = r.json()
    per_page = 200
    page_limit = (data["totalHits"] // per_page) + 1
    return page_limit
count =1
#write the terms you want to find 
terms_to_find = ["bed","fraud"]
for terms in terms_to_find:    
    
    download_dir = "H:/" + terms #put your directory file name here
    if not os.path.exists(download_dir):
        os.mkdir(download_dir) 
    page_number = find_pages_limit(terms)
    print(terms)
    print(page_number)
  
    for page in range(1,page_number+1): 
        print(terms)
        print(page) 
        url_req = (f"https://pixabay.com/api/?key=21967882-1b9779f8b2e624ded8ba912a1&page={page}&q={terms}&per_page=200&image_type=photo")
        print(url_req)
        r = requests.get(url_req)
        data = r.json()
        print(len(data['hits']))
        for f,i in enumerate(data['hits']):
            print(f)
            name = str(data['hits'][f]['id']) + "page_no"+str(page) + str(count)
            print(name)
            count = count + 1
            url = data['hits'][f]['webformatURL']
            filepath = f"{os.path.join(os.path.realpath(os.getcwd()),download_dir,name)}.jpg"
            print(name)
            print(url)
            #if not os.path.exists(filepath):
            with open(filepath,"wb") as f:
                f.write(requests.request("GET",url).content)
                print(f)
    r.close()
```
